```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv
from urllib.parse import quote_plus

# Load environment variables
load_dotenv(r"G:\My Drive\User Retention & Business Decision Impact Analysis Project\Olist Project Notebook\.env")

DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")

# Encode password safely
password = quote_plus(DB_PASSWORD)

# Dataset path
path = r"G:\My Drive\User Retention & Business Decision Impact Analysis Project\Brazilian E-Commerce Public Dataset by Olist"

# Create PostgreSQL engine
engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{password}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

# Loop through all CSV files
for file in os.listdir(path):
    if file.endswith(".csv"):
        table_name = "raw_" + file.replace("olist_", "").replace("_dataset", "").replace(".csv", "")

        df = pd.read_csv(os.path.join(path, file), low_memory=False)
        df.columns = df.columns.str.lower().str.strip()

        df.to_sql(
            table_name,
            engine,
            if_exists="replace",
            index=False,
            method="multi",
            chunksize=5000
        )

        print(f"Loaded {table_name}")

print("All CSV files loaded into PostgreSQL")
```
